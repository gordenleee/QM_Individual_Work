## 1. Loading and Inspecting Data

```python
import pandas as pd

# Load the data
file_path = 'data/google_activity_by_London_Borough.csv'
data = pd.read_csv(file_path)

file_path = 'data/phe_deaths_london_boroughs.csv'
covid_data = pd.read_csv(file_path)

# 左合并 mobility_data 和 covid_data，基于'date'和'area_code'
data = pd.merge(data, 
                         covid_data[['date', 'area_code', 'new_deaths']], 
                         left_on=['date', 'area_code'], 
                         right_on=['date', 'area_code'], 
                         how='left')

# 缺失的'new_deaths'数据填充为0
data['new_deaths'] = data['new_deaths'].fillna(0)

# Display the first few rows of the dataframe
data.head()
```

## 2. Data Pre-processing

```python
# 1. Remove 'Unnamed: 0'
data = data.drop(columns=['Unnamed: 0'])

# 2. Convert 'date' to datetime
data['date'] = pd.to_datetime(data['date'])

# 3. Check for missing values
missing_values = data.isnull().sum()

# 4. Check data types
data_types = data.dtypes

missing_values, data_types
```

## 3. Handling Missing Values

```python
# Handling missing values with forward fill, then backward fill for remaining NAs
data_filled = data.groupby('area_name').apply(lambda group: group.ffill().bfill())

# Re-checking for missing values after imputation
missing_values_after = data_filled.isnull().sum()
missing_values_after
```

## 4. Data Validation

```python
# Validate data: Checking the range and distribution of dates and percentage changes
date_range = (data_filled['date'].min(), data_filled['date'].max())
retail_range = (data_filled['retail_and_recreation_percent_change_from_baseline'].min(), 
                data_filled['retail_and_recreation_percent_change_from_baseline'].max())
grocery_range = (data_filled['grocery_and_pharmacy_percent_change_from_baseline'].min(), 
                 data_filled['grocery_and_pharmacy_percent_change_from_baseline'].max())
parks_range = (data_filled['parks_percent_change_from_baseline'].min(), 
               data_filled['parks_percent_change_from_baseline'].max())
transit_range = (data_filled['transit_stations_percent_change_from_baseline'].min(), 
                 data_filled['transit_stations_percent_change_from_baseline'].max())
workplaces_range = (data_filled['workplaces_percent_change_from_baseline'].min(), 
                    data_filled['workplaces_percent_change_from_baseline'].max())
residential_range = (data_filled['residential_percent_change_from_baseline'].min(), 
                     data_filled['residential_percent_change_from_baseline'].max())

date_range, retail_range, grocery_range, parks_range, transit_range, workplaces_range, residential_range
```

## 5. Time Series Analysis

```python
import matplotlib.pyplot as plt

# Select a few representative boroughs for visualization
sample_boroughs = ['Greenwich', 'Hackney', 'Westminster', 'Croydon']

# Setting up the plots
fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(15, 20), sharex=True)

# Plotting time series for each selected borough
for i, borough in enumerate(sample_boroughs):
    borough_data = data_filled[data_filled['area_name'] == borough]
    axes[i].plot(borough_data['date'], borough_data['retail_and_recreation_percent_change_from_baseline'], label='Retail & Recreation')
    axes[i].plot(borough_data['date'], borough_data['grocery_and_pharmacy_percent_change_from_baseline'], label='Grocery & Pharmacy')
    axes[i].plot(borough_data['date'], borough_data['parks_percent_change_from_baseline'], label='Parks')
    axes[i].plot(borough_data['date'], borough_data['transit_stations_percent_change_from_baseline'], label='Transit Stations')
    axes[i].plot(borough_data['date'], borough_data['workplaces_percent_change_from_baseline'], label='Workplaces')
    axes[i].plot(borough_data['date'], borough_data['residential_percent_change_from_baseline'], label='Residential', linestyle='--')
    axes[i].set_title(f"Mobility Trends in {borough}")
    axes[i].legend()
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```

## 7. Cluster Analysis

```python
# Drop rows that still have missing values after forward/backward fill
data_filled = data_filled.dropna().reset_index(drop=True)

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# 假设你已经有了combined_data，其中包括了date, area_code, new_deaths以及其他移动性特征

# 定义时间段（基于政策变化的关键日期）
time_periods = [
    ('2020-02-15', '2020-03-23'),
    ('2020-03-24', '2020-07-07'), 
    ('2020-07-08', '2020-10-30'), 
    ('2020-10-31', '2020-12-01'),
    ('2020-12-02', '2020-12-31'),
    ('2021-01-01', '2021-12-15'),
    ('2021-12-16', '2021-12-31'),
    ('2022-01-01', '2022-01-18'),
    ('2022-01-19', '2022-03-17'),
    ('2022-03-18', '2022-10-15'),
]

# 定义用于聚类的旧特征和新特征
old_features = [
    'retail_and_recreation_percent_change_from_baseline', 
    'grocery_and_pharmacy_percent_change_from_baseline',
    'parks_percent_change_from_baseline',
    'transit_stations_percent_change_from_baseline',
    'workplaces_percent_change_from_baseline',
    'residential_percent_change_from_baseline'
]
new_feature = 'new_deaths'
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import pandas as pd

# Dictionary to store the cluster labels and scaled data for each time period
cluster_assignments = {}
scaled_data_dict = {}  # To store the scaled features for silhouette score calculation

# Perform clustering for each time period and store results
for start_date, end_date in time_periods:
    # Select current time period data
    period_data = data_filled[(data_filled['date'] >= start_date) & (data_filled['date'] <= end_date)]

    # Calculate mean of old features and sum of new features for each area
    period_agg = period_data.groupby('area_code').agg({**{col: 'mean' for col in old_features}, new_feature: 'sum'}).reset_index()

    # Standardize the features
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(period_agg.iloc[:, 1:])  # Exclude the area_code column

    # Determine the optimal number of clusters using Elbow or Silhouette Method
    # Initialize variables to store the results
    silhouette_scores = []
    inertia_scores = []
    range_n_clusters = list(range(2, 11))  # for instance, testing from 2 to 10 clusters

    for n_clusters in range_n_clusters:
        kmeans = KMeans(n_clusters=n_clusters, random_state=1)
        cluster_labels = kmeans.fit_predict(scaled_features)

        # For silhouette score
        silhouette_avg = silhouette_score(scaled_features, cluster_labels)
        silhouette_scores.append(silhouette_avg)

        # For Elbow method (using inertia)
        inertia_scores.append(kmeans.inertia_)

    # Determine the optimal number of clusters
    # This can be done by looking for a plateau or "elbow" in the plot of inertia, or the highest silhouette score
    # For simplicity, this example uses silhouette score; you can use the elbow method by plotting inertia and visually inspecting the plot
    optimal_n_clusters = range_n_clusters[silhouette_scores.index(max(silhouette_scores))]

    # Perform KMeans clustering with the optimal number of clusters
    optimal_kmeans = KMeans(n_clusters=optimal_n_clusters, random_state=1)
    period_agg['Cluster'] = optimal_kmeans.fit_predict(scaled_features)

    # 输出每个集群的行政区域
    print(f"Clusters for period {start_date} to {end_date}:")
    boroughs_by_cluster = period_agg.groupby('Cluster')['area_code'].unique()
    print(boroughs_by_cluster)
    print("\n")
    
    # Store the cluster labels and scaled features in the dictionaries
    cluster_assignments[(start_date, end_date)] = period_agg['Cluster'].tolist()
    scaled_data_dict[(start_date, end_date)] = scaled_features

# 计算并打印轮廓系数
for period, clusters in cluster_assignments.items():
    silhouette_avg = silhouette_score(scaled_data_dict[period], clusters)
    print(f"Silhouette Coefficient for period {period}: {silhouette_avg}")

# 分析区域在不同时间段的聚类分配
area_cluster_time_series = pd.DataFrame(index=data_filled['area_code'].unique())

for period, clusters in cluster_assignments.items():
    area_cluster_time_series[period] = clusters

# 创建交叉表来分析区域与聚类的关系
all_periods = sorted(cluster_assignments.keys())
for i in range(len(all_periods) - 1):
    current_period = all_periods[i]
    next_period = all_periods[i+1]
    ct = pd.crosstab(area_cluster_time_series[current_period], 
                     area_cluster_time_series[next_period])
    print(f"Cross-tabulation between {current_period} and {next_period}:")
    print(ct)
    print("\n")

# 注意：你需要确保data_filled['area_code']中没有重复值，并且cluster_assignments中的聚类结果与这些区域代码是一致的。
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# 获取聚类中心
centers = kmeans.cluster_centers_

# 逆标准化聚类中心
original_centers = scaler.inverse_transform(centers)

# 创建一个数据框来存放和展示聚类中心
cluster_centers_df = pd.DataFrame(original_centers, columns=old_features + [new_feature])

# 打印每个聚类的中心
print(cluster_centers_df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Make sure cluster_centers_df is a DataFrame and has correct columns
print(type(cluster_centers_df))  # Should print <class 'pandas.core.frame.DataFrame'>

# Calculate the number of variables we're plotting.
num_vars = len(cluster_centers_df.columns)

# Create a list of angles, which are evenly spaced around a circle.
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # Repeat the first value to close the circle

# Ensure the DataFrame is "cyclic" by appending the first row to the end.
cyclic_cluster_centers = pd.concat([cluster_centers_df, cluster_centers_df.iloc[:1]])

# Plotting
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
for idx, row in cyclic_cluster_centers.iterrows():
    # Ensure the plot is cyclic and starts at the top
    data = row.tolist()
    data += data[:1]  # Ensure the polygon is closed by repeating the first value
    ax.plot(angles, data, 'o-', linewidth=2, label="Cluster {}".format(idx))
    ax.fill(angles, data, alpha=0.25)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(cluster_centers_df.columns)

# Remove y-tick labels and add legend
ax.set_yticklabels([])
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1.05))
plt.show()
```

```python
import pandas as pd
from collections import Counter

# Create a DataFrame to hold the cluster time series for each area
area_cluster_time_series = pd.DataFrame(index=data_filled['area_code'].unique())

# Fill the DataFrame with cluster assignments for each time period
for period, clusters in cluster_assignments.items():
    # Assume that 'area_code' is the identifier for areas
    # and that it's aligned with the order of clusters in 'clusters'
    area_cluster_time_series[period] = clusters

# Now, area_cluster_time_series contains the cluster number for each area in each time period
print("Cluster Time Series for Each Area:")
print(area_cluster_time_series.head())

# Analyze transitions: count how often areas move between specific clusters from one period to the next
transition_counts = {period: Counter() for period in area_cluster_time_series.columns[1:]}

for i, current_period in enumerate(area_cluster_time_series.columns[:-1]):
    next_period = area_cluster_time_series.columns[i + 1]
    transitions = zip(area_cluster_time_series[current_period], area_cluster_time_series[next_period])
    
    for from_cluster, to_cluster in transitions:
        transition_counts[next_period][(from_cluster, to_cluster)] += 1

# Print the transition counts to understand how areas move between clusters
for period, counts in transition_counts.items():
    print(f"Transition counts from previous period to {period}:")
    for transition, count in counts.items():
        print(f"From Cluster {transition[0]} to Cluster {transition[1]}: {count} areas")
    print("\n")

# Further Analysis:
# You might want to visualize these transitions using network graphs, heatmaps, or other visualization tools
# to better understand the movement patterns and possibly correlate them with external events or policies.
```

```python
transition_matrices = {}

# 绘制每个时间段的转移矩阵热图
area_cluster_time_series = pd.DataFrame(index=data_filled['area_code'].unique())
for period, clusters in cluster_assignments.items():
    area_cluster_time_series[period] = clusters

# 计算转移矩阵并绘制热图
for i in range(len(time_periods) - 1):
    previous_period = time_periods[i]
    current_period = time_periods[i + 1]
    ct = pd.crosstab(area_cluster_time_series[previous_period], area_cluster_time_series[current_period])
    transition_matrices[(previous_period, current_period)] = ct
    plt.figure(figsize=(10, 8))
    sns.heatmap(ct, annot=True, fmt="d", cmap="YlGnBu")
    plt.title(f"Transition from {previous_period} to {current_period}")
    plt.ylabel('From Cluster (Previous Period)')
    plt.xlabel('To Cluster (Current Period)')
    plt.show()
```
